## <font color="Red">1.1 Calculate gray image from color image</font>
<img src="IP01_1.png" width="600px">

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imgBGR = cv2.imread("Commons.jpeg", cv2.IMREAD_COLOR)
img = imgBGR[:,:,::-1]

# Image size and informstion
nr, nc = img.shape[:2]
gray = np.zeros((nr,nc),dtype=np.uint8)

#-------------- Code Here ------------------#
for x in range(nr):
  for y in range(nc):
    r,g,b=img[x,y]
    gray[x,y]=0.299*r+0.587*g+0.114*b
#-------------- Stop Here ------------------#

plt.figure(figsize=(12, 10))
plt.subplot(1,2,1)
plt.title('Image color')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Image gray')
plt.imshow(gray, cmap=plt.cm.gray)
```

## <font color="Red">1.2 Calculate Cb and Cr</font>

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

imgBGR = cv2.imread("Commons.jpeg", cv2.IMREAD_COLOR)
img = imgBGR[:,:,::-1]

# Image size and informstion
nr, nc = img.shape[:2]
Cb = np.zeros((nr,nc),dtype=np.uint8)
Cr = np.zeros((nr,nc),dtype=np.uint8)
#-------------- Code Here ------------------#
for x in range(nr):
  for y in range(nc):
    r,g,b=img[x,y]
    Cb[x,y]=-0.169*r-0.331*g+0.5*b+128
    Cr[x,y]=0.5*r-0.419*g-0.081*b+128
#-------------- Stop Here ------------------#
plt.figure(figsize=(12, 10))
plt.subplot(1,3,1)
plt.title('Image color')
plt.imshow(img)
plt.subplot(1,3,2)
plt.title('Image Cb')
plt.imshow(Cb)
plt.subplot(1,3,3)
plt.title('Image Cr')
plt.imshow(Cr)
```

## <font color="Red"> 1.4 Embed text into images and show as follows. </font>

## <font color="Red"> 1.6 Please create the following two images by Python</font><p>

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math

img = np.zeros( [ 512, 512, 1 ], dtype = 'uint8' )

#-------------- Code Here ------------------#
block_size=64
for i in range(0,img.shape[0],block_size):
    for j in range(0,img.shape[1],block_size):
        if (i//block_size)%2==(j//block_size)%2:
            img[i:i+block_size,j:j+block_size]=0
        else:
            img[i:i+block_size,j:j+block_size]=255
#-------------- Stop Here ------------------#


plt.figure(figsize=(12, 10))
plt.title('original image')
plt.imshow(img,cmap=plt.cm.gray)
```

## <font color="Red"> 1.7 Please create the following two images by Python</font><p>
Please use the folloeing bit extraction method to acquire the bit plane image: <p>
val = 63 <p>
bit_plane = 0b00000001 <p>
val & bit_plane = 1 <p>
<img src="IP01_h3.png" width="800px">    

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def bit_plane(f, bit):
#-------------- Code Here ------------------#
  bp = np.bitwise_and(f, 1 << bit)
  bp = bp >> bit
#-------------- Stop Here ------------------#
  return bp
img = cv2.imread("Lenna.bmp", -1)
bit = 7
bitplane1 =  bit_plane(img, bit)

plt.figure(figsize=(12, 10))
plt.subplot(2,2,1)
plt.title('source image')
plt.imshow(img,cmap=plt.cm.gray)
plt.subplot(2,2,2)
plt.title('image of bit plane '+ str(bit))
plt.imshow(bitplane1,cmap=plt.cm.gray)
```
